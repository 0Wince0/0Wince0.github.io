```python
import pandas as pd
from glob import glob
import os
import re
```

# 넷플릭스에서 가져온 대본 불러오기

```python
file_path = '.\\data\\'
#data 폴더에 대본들 저장

file_name_os=os.listdir(file_path)
file_name_glob=glob(file_path+'*')
file_name_glob
```

# 전체 대본 중 도라에몽 대사만을 추출

```python
df_list = []
for fn in file_name_os :
    df_list.append(pd.read_excel(file_path + fn))
list_data = pd.concat(df_list)
list_data
#읽은 엑셀파일 순서대로 열 방향으로 concat
```

```python
대본 =list_data.drop(['Time','Translation','Transliteration'], axis=1)
대본
#필요없는 열 제거
```

```python
도라에몽=[]
for ii in 대본['Subtitle'] :
    if '(도라에몽)' in ii :
        도라에몽.append(ii)
    else:
        pass
```

```python
도라에몽
```

```python
도라에몽2=[]
for ii in 도라에몽:
    x=ii.split('-')
    for i in x:
        if ('(도라에몽)' or ' (도라에몽)') in i:
             도라에몽2.append(i)
```

```python
도라에몽2
```

```python
도라에몽3=[]
for ii in 도라에몽2:
    x= re.sub(r'\[[^]]*\]', '', ii)
    도라에몽3.append(x)
```

```python
도라에몽4=[]
for ii in 도라에몽3:
    if (ii =='(도라에몽)' or ii == ' (도라에몽)'):
        pass
    else:
        도라에몽4.append(ii)
```

```python
도라에몽4
```

```python
완성본 = pd.DataFrame(도라에몽4, columns=['대사'])
```

```python
완성본.to_csv('완성본.csv',encoding='cp949')
```
