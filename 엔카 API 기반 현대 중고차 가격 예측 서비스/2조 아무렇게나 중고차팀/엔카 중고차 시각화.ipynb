```python
from scipy.stats import ttest_ind
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rc
import seaborn as sns
%matplotlib inline

# üîπ ÌïúÍ∏Ä Ìè∞Ìä∏ ÏÑ§Ï†ï (Íπ®Ïßê Î∞©ÏßÄ)
plt.rc('font', family='Malgun Gothic')

# Í∑∏ÎûòÌîÑÏóê ÎßàÏù¥ÎÑàÏä§ Í∏∞Ìò∏ Íπ®ÏßÄÎäî Î¨∏Ï†ú Ìï¥Í≤∞
plt.rcParams['axes.unicode_minus'] = False

# ÌååÏùº Í≤ΩÎ°ú ÏÑ§Ï†ï (ÌïòÎÇòÏùò ÌååÏùºÏóêÏÑú Îç∞Ïù¥ÌÑ∞ Î∂àÎü¨Ïò§Í∏∞)
file_path = "data/·Ñâ·Öµ·ÑÄ·Ö°·Ü®·Ñí·Ö™_·Ñí·Öß·Ü´·ÑÉ·Ö¢_·ÑÄ·Öµ·Ñá·Ö©·Ü´_·Ñå·ÖÆ·Üº·Ñá·Ö©·Ü®·Ñå·Ö¶·ÑÄ·Ö•_·Ñé·Ö°·ÑÖ·Ö£·Üº·Ñá·Ö•·Ü´·Ñí·Ö©_·Ñá·Ö©·Ñí·Öß·Ü∑·Ñã·Öµ·ÑÖ·Öß·Ü®.csv"

# Îç∞Ïù¥ÌÑ∞ Î∂àÎü¨Ïò§Í∏∞
df = pd.read_csv(file_path)
```

# Ï∞®ÎüâÏó∞ÏãùÏóê Îî∞Î•∏ Í∞ÄÍ≤© Î∂ÑÌè¨

#### Ï∞®ÎüâÏó∞ÏãùÏù¥ Í∞ÄÍ≤© Î≥ÄÎèôÏóê ÎØ∏ÏπòÎäî ÏòÅÌñ•ÏùÑ Î≥¥Ïó¨Ï£ºÍ≥†ÏûàÎã§.

```python
# Î≥ÄÏàò Ïù¥Î¶Ñ ÏàòÏ†ï
df['Ïó∞Ïãù'] = df['Ïó∞Ïãù'].astype(int)
df['Ïó∞Ïãù'] = df['Ïó∞Ïãù'].astype(str).str[:4].astype(int)  # Ïó∞ÏãùÏùÑ Ïà´ÏûêÎ°ú Î≥ÄÌôò
df = df.sort_values(by='Ïó∞Ïãù')

# Ïó∞ÏãùÎ≥Ñ ÌèâÍ∑† Í∞ÄÍ≤© Í≥ÑÏÇ∞
mean_prices = df.groupby('Ïó∞Ïãù')['Í∞ÄÍ≤©'].mean().reset_index()

# Í∑∏ÎûòÌîÑ Í∑∏Î¶¨Í∏∞
plt.figure(figsize=(15, 8))
sns.scatterplot(x='Ïó∞Ïãù', y='Í∞ÄÍ≤©', data=df, alpha=0.5, color='lightblue', label="ÏÇ∞Ï†êÎèÑ")

# ÏÑ† Í∑∏ÎûòÌîÑ Ï∂îÍ∞Ä (Îπ®Í∞ÑÏÉâ)
plt.plot(mean_prices['Ïó∞Ïãù'], mean_prices['Í∞ÄÍ≤©'], color='red', linewidth=2, label="ÌèâÍ∑† Í∞ÄÍ≤© Ï∂îÏÑ∏ÏÑ†")

# xÏ∂ï ÎààÍ∏à ÏÑ§Ï†ï (5ÎÖÑ Îã®ÏúÑ)
years = np.arange(df['Ïó∞Ïãù'].min(), df['Ïó∞Ïãù'].max() + 1, 5)
plt.xticks(years, rotation=90)

# Ï†úÎ™© Î∞è ÎùºÎ≤® ÏÑ§Ï†ï
plt.title('Ïó∞ÏãùÏóê Îî∞Î•∏ Í∞ÄÍ≤© Î∂ÑÌè¨')
plt.xlabel('Ïó∞Ïãù')
plt.ylabel('Í∞ÄÍ≤© (Îßå Ïõê)')
plt.legend()

plt.show()
```

# Ï∞®ÎüâÏãúÎ¶¨Ï¶àÎ≥Ñ Í∞ÄÍ≤© Î∂ÑÌè¨

#### Ï∞®Îüâ ÏãúÎ¶¨Ï¶àÎ≥Ñ ex) ÏïÑÎ∞òÎñº, ÏÜåÎÇòÌÉÄ, ÏÇ∞ÌÉÄÌéò Îì±Îì± Î≥Ñ Í∞ÅÍ∞Å Í∞ÄÍ≤©Ïù¥ ÏñºÎßàÏù∏ÏßÄÎ•º ÎÇòÌÉÄÎÇ¥Ï£ºÍ≥†ÏûàÎã§. 

```python
# Ï°∞Í±¥Ïóê ÎßûÎäî Îç∞Ïù¥ÌÑ∞ ÌïÑÌÑ∞ÎßÅ
df_filtered = df[(df['ÎÇ¥Ï∞®ÌîºÌï¥'] == 0) & 
                 (df['ÌÉÄÏ∞®Í∞ÄÌï¥'] == 0) & 
                 (df['Ï†ÑÏ≤¥Ïπ®ÏàòÌöüÏàò'] == 0)].copy()

# Ï†ÄÍ∞Ä/Í≥†Í∏â Î™®Îç∏ Î∂ÑÎ•ò
total_mean = df_filtered['Í∞ÄÍ≤©'].mean()
model_avg = df_filtered.groupby('Î™®Îç∏ Ï¢ÖÎ•ò')['Í∞ÄÍ≤©'].mean().reset_index()
low_price_threshold = model_avg['Í∞ÄÍ≤©'].quantile(0.25)
high_price_threshold = model_avg['Í∞ÄÍ≤©'].quantile(0.75)

model_avg['Í∑∏Î£π'] = np.where(
    model_avg['Í∞ÄÍ≤©'] <= low_price_threshold, 'Ï†ÄÍ∞Ä', 
    np.where(model_avg['Í∞ÄÍ≤©'] >= high_price_threshold, 'Í≥†Í∏â', 'Ï§ëÍ∞Ñ')
)

df_filtered = pd.merge(df_filtered, model_avg[['Î™®Îç∏ Ï¢ÖÎ•ò', 'Í∑∏Î£π']], on='Î™®Îç∏ Ï¢ÖÎ•ò')

# Í∑∏Î£πÎ≥Ñ ÌÜµÍ≥ÑÏπò ÎπÑÍµê
group_stats = df_filtered.groupby('Í∑∏Î£π')['Í∞ÄÍ≤©'].agg(['mean', 'var', 'std', 'count'])
print("Ï†ÄÍ∞Ä/Í≥†Í∏â Î™®Îç∏ ÌÜµÍ≥Ñ ÎπÑÍµê:")
print(group_stats.round(2))

# ÏãúÍ∞ÅÌôî
plt.figure(figsize=(16, 10))

# Î™®Îç∏ Ï†ïÎ†¨ (ÌèâÍ∑† Í∞ÄÍ≤© Í∏∞Ï§Ä)
model_order = df_filtered.groupby('Î™®Îç∏ Ï¢ÖÎ•ò')['Í∞ÄÍ≤©'].mean().sort_values().index.tolist()

# Î™®Îç∏Î≥Ñ ÌÜµÍ≥ÑÍ∞í Í≥ÑÏÇ∞ (Î∂ÑÏÇ∞, ÌëúÏ§ÄÌé∏Ï∞®, Q3)
model_stats = df_filtered.groupby('Î™®Îç∏ Ï¢ÖÎ•ò')['Í∞ÄÍ≤©'].agg(['var', 'std', lambda x: x.quantile(0.75)]).reset_index()
model_stats.columns = ['Î™®Îç∏ Ï¢ÖÎ•ò', 'Î∂ÑÏÇ∞', 'ÌëúÏ§ÄÌé∏Ï∞®', 'Q3']

# Î∞ïÏä§ÌîåÎ°Ø Í∑∏Î¶¨Í∏∞
ax = sns.boxplot(
    x='Î™®Îç∏ Ï¢ÖÎ•ò', 
    y='Í∞ÄÍ≤©', 
    data=df_filtered, 
    order=model_order,
    palette={'Ï†ÄÍ∞Ä': '#1f77b4', 'Í≥†Í∏â': '#ff7f0e', 'Ï§ëÍ∞Ñ': '#2ca02c'},
    hue='Í∑∏Î£π',
    showfliers=False
)

# Î∂ÑÏÇ∞Í≥º ÌëúÏ§ÄÌé∏Ï∞®Î•º Î∞ïÏä§ ÏÉÅÎã®Ïóê ÌëúÏãú
for i, model in enumerate(model_order):
    stats = model_stats[model_stats['Î™®Îç∏ Ï¢ÖÎ•ò'] == model]
    var = stats['Î∂ÑÏÇ∞'].values[0]
    std = stats['ÌëúÏ§ÄÌé∏Ï∞®'].values[0]
    q3 = stats['Q3'].values[0]
    
    ax.text(
        i, 
        q3 + 200,  # Q3 Í∏∞Ï§ÄÏúºÎ°ú 200ÎßåÏõê ÏúÑÏóê ÌëúÏãú
        f"Var: {var:.2f}\nStd: {std:.2f}",
        ha='center',
        va='bottom',
        fontsize=10,
        bbox=dict(facecolor='white', alpha=0.9, edgecolor='gray', boxstyle='round')
    )

# ÌèâÍ∑†Í∞í Í≥ÑÏÇ∞ Î∞è Ï∂îÍ∞Ä
means = df_filtered.groupby('Î™®Îç∏ Ï¢ÖÎ•ò')['Í∞ÄÍ≤©'].mean()
for i, model in enumerate(model_order):
    mean_price = means[model]
    
    # ÌèâÍ∑†Í∞í ÎßàÏª§ Ï∂îÍ∞Ä (Îπ®Í∞Ñ Ï†ê)
    ax.scatter(i, mean_price, color='red', zorder=5, s=100)
    
    # ÌèâÍ∑†Í∞í ÌÖçÏä§Ìä∏ Ï∂îÍ∞Ä (ÍµµÏùÄ Ìù∞ÏÉâ Í∏ÄÏûê)
    ax.text(
        i, mean_price - 150, f"{mean_price:.0f}",  # yÏ¢åÌëúÎ•º ÌèâÍ∑†Í∞íÎ≥¥Îã§ 100Îßå Ïõê ÎÇÆÍ≤å ÏÑ§Ï†ï
        ha='center', va='top', fontsize=10,
        color='white', fontweight='bold',
        # bbox=dict(facecolor='black', edgecolor='none', boxstyle='round,pad=0.3')
    )


# Í∑∏ÎûòÌîÑ ÏÑ§Ï†ï
plt.title("ÏãúÎ¶¨Ï¶àÎ≥Ñ Í∞ÄÍ≤© Î∂ÑÌè¨ (Ï†ÄÍ∞Ä vs Í≥†Í∏â)", fontsize=18, pad=20)
plt.xlabel('ÏãúÎ¶¨Ï¶à Ï¢ÖÎ•ò', fontsize=14, labelpad=15)
plt.ylabel('Í∞ÄÍ≤© (ÎßåÏõê)', fontsize=14, labelpad=15)
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', ls=':')
plt.legend(title='Î™®Îç∏ Í∑∏Î£π', loc='upper left', fontsize=12)

plt.show()
```

# Ï£ºÌñâÍ±∞Î¶¨Ïóê Îî∞Î•∏ ÌèâÍ∑† Í∞ÄÍ≤©Î≥ÄÌôî

#### Ï∞®Îüâ Ï£ºÌñâÍ±∞Î¶¨ Ï¶ùÍ∞ÄÏóê Îî∞Î•∏ Í∞ÄÍ≤©Î≥ÄÌôîÎ•º ÌèâÍ∑†ÏùÑ ÎÇ¥ÏÑú Í∑∏ÎûòÌîÑÎ°ú ÌôïÏù∏ÌïòÍ≥†ÏûàÎã§.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Í∑∏ÎûòÌîÑÏóê ÌïúÍ∏Ä ÏÑ§Ï†ï
plt.rcParams['font.family'] = 'Malgun Gothic'

# Í∑∏ÎûòÌîÑÏóê ÎßàÏù¥ÎÑàÏä§ Í∏∞Ìò∏ Íπ®ÏßÄÎäî Î¨∏Ï†ú Ìï¥Í≤∞
plt.rcParams['axes.unicode_minus'] = False

# ÌååÏùº Í≤ΩÎ°ú ÏÑ§Ï†ï
file_path = "data/·Ñâ·Öµ·ÑÄ·Ö°·Ü®·Ñí·Ö™_·Ñí·Öß·Ü´·ÑÉ·Ö¢_·ÑÄ·Öµ·Ñá·Ö©·Ü´_·Ñå·ÖÆ·Üº·Ñá·Ö©·Ü®·Ñå·Ö¶·ÑÄ·Ö•_·Ñé·Ö°·ÑÖ·Ö£·Üº·Ñá·Ö•·Ü´·Ñí·Ö©_·Ñá·Ö©·Ñí·Öß·Ü∑·Ñã·Öµ·ÑÖ·Öß·Ü®.csv"

# Îç∞Ïù¥ÌÑ∞ Î∂àÎü¨Ïò§Í∏∞
df = pd.read_csv(file_path)

# Ï°∞Í±¥Ïóê ÎßûÎäî Îç∞Ïù¥ÌÑ∞ ÌïÑÌÑ∞ÎßÅ
df_filtered = df[(df['ÎÇ¥Ï∞®ÌîºÌï¥'] == 0) & 
                 (df['ÌÉÄÏ∞®Í∞ÄÌï¥'] == 0) & 
                 (df['Ï†ÑÏ≤¥Ïπ®ÏàòÌöüÏàò'] == 0)].copy()

# Ï£ºÌñâÍ±∞Î¶¨Î•º 10000km Îã®ÏúÑÎ°ú Í∑∏Î£πÌôî
df_filtered['Ï£ºÌñâÍ±∞Î¶¨_Í∑∏Î£π'] = ((df_filtered['Ï£ºÌñâÍ±∞Î¶¨'] // 10000) + 1) * 10000

# Î™®Îç∏Î≥ÑÎ°ú Í∑∏Î£πÌôîÌïòÍ≥† ÌèâÍ∑† Í∞ÄÍ≤© Í≥ÑÏÇ∞
grouped = df_filtered.groupby(['Ï£ºÌñâÍ±∞Î¶¨_Í∑∏Î£π'])['Í∞ÄÍ≤©'].mean().reset_index()

# Îç∞Ïù¥ÌÑ∞ ÌÉÄÏûÖ Î≥ÄÌôò
grouped.loc[:, 'Ï£ºÌñâÍ±∞Î¶¨_Í∑∏Î£π'] = grouped['Ï£ºÌñâÍ±∞Î¶¨_Í∑∏Î£π'].astype(int)
grouped.loc[:, 'Í∞ÄÍ≤©'] = grouped['Í∞ÄÍ≤©'].astype(float)

# Ï£ºÌñâÍ±∞Î¶¨ Í∑∏Î£πÏù¥ 400,000km Ïù¥ÌïòÏù∏ Îç∞Ïù¥ÌÑ∞ ÌïÑÌÑ∞ÎßÅ
grouped = grouped[grouped['Ï£ºÌñâÍ±∞Î¶¨_Í∑∏Î£π'] <= 400000]

plt.figure(figsize=(12, 8))

# Í∑∏ÎûòÌîÑ Í∑∏Î¶¨Í∏∞ (Í∏∞Î≥∏ ÌååÎûÄÏÉâÏúºÎ°ú ÏÑ§Ï†ï)
plt.bar(grouped['Ï£ºÌñâÍ±∞Î¶¨_Í∑∏Î£π'].astype(str), grouped['Í∞ÄÍ≤©'], color='#1f77b4')  # Í∏∞Î≥∏ ÌååÎûÄÏÉâ ÏÑ§Ï†ï

# yÏ∂ï Î≤îÏúÑ ÏÑ§Ï†ï
plt.ylim(0, grouped['Í∞ÄÍ≤©'].max() * 1.1)

# Îç∞Ïù¥ÌÑ∞ Ìè¨Ïù∏Ìä∏ ÌëúÏãú
for i, v in enumerate(grouped['Í∞ÄÍ≤©']):
    plt.text(i, v, f'{v:.0f}', ha='center', va='bottom', fontsize=8)

plt.title(f"Ï£ºÌñâÍ±∞Î¶¨Ïóê Îî∞Î•∏ ÌèâÍ∑† Í∞ÄÍ≤© Î≥ÄÌôî", fontsize=16, pad = 20)
plt.xlabel('Ï£ºÌñâÍ±∞Î¶¨ (km)', fontsize=14, labelpad=15)
plt.ylabel('ÌèâÍ∑† Í∞ÄÍ≤© (ÎßåÏõê)', fontsize=14, labelpad=15)
plt.ylim(0, 5000)

# xÏ∂ï Î†àÏù¥Î∏î ÏÑ§Ï†ï Ïãú Ï†ïÏàòÌòïÏúºÎ°ú Î≥ÄÌôòÌïòÏó¨ ÏÜåÏàòÏ†ê Ï†úÍ±∞
plt.xticks(ticks=range(len(grouped)), labels=[f"{int(x):,}" for x in grouped['Ï£ºÌñâÍ±∞Î¶¨_Í∑∏Î£π']], rotation=45)

plt.tight_layout()

# Í∑∏ÎûòÌîÑÎ•º JPG ÌååÏùºÎ°ú Ï†ÄÏû•
plt.savefig('bar_average_price_by_mileage.jpg', format='jpg', dpi=300)

plt.show()
```

#### Ï∞®Îüâ Ï£ºÌñâÍ±∞Î¶¨ Ï¶ùÍ∞ÄÏóê Îî∞Î•∏ Í∞ÄÍ≤©Î≥ÄÌôîÎ•º ÌèâÍ∑†ÏùÑ ÎÇ¥ÏÑú ÏÇ∞Ìè¨ÎèÑ Î∞è ÏÑ†ÌòïÌöåÍ∑Ä Í∑∏ÎûòÌîÑÎ°ú ÌôïÏù∏ÌïòÍ≥†ÏûàÎã§.

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Í∑∏ÎûòÌîÑÏóê ÌïúÍ∏Ä ÏÑ§Ï†ï
plt.rcParams['font.family'] = 'Malgun Gothic'

# Í∑∏ÎûòÌîÑÏóê ÎßàÏù¥ÎÑàÏä§ Í∏∞Ìò∏ Íπ®ÏßÄÎäî Î¨∏Ï†ú Ìï¥Í≤∞
plt.rcParams['axes.unicode_minus'] = False

# ÌååÏùº Í≤ΩÎ°ú ÏÑ§Ï†ï
file_path = "data/·Ñâ·Öµ·ÑÄ·Ö°·Ü®·Ñí·Ö™_·Ñí·Öß·Ü´·ÑÉ·Ö¢_·ÑÄ·Öµ·Ñá·Ö©·Ü´_·Ñå·ÖÆ·Üº·Ñá·Ö©·Ü®·Ñå·Ö¶·ÑÄ·Ö•_·Ñé·Ö°·ÑÖ·Ö£·Üº·Ñá·Ö•·Ü´·Ñí·Ö©_·Ñá·Ö©·Ñí·Öß·Ü∑·Ñã·Öµ·ÑÖ·Öß·Ü®.csv"

# Îç∞Ïù¥ÌÑ∞ Î∂àÎü¨Ïò§Í∏∞
df = pd.read_csv(file_path)

# Ï°∞Í±¥Ïóê ÎßûÎäî Îç∞Ïù¥ÌÑ∞ ÌïÑÌÑ∞ÎßÅ
df_filtered = df[(df['ÎÇ¥Ï∞®ÌîºÌï¥'] == 0) & 
                 (df['ÌÉÄÏ∞®Í∞ÄÌï¥'] == 0) & 
                 (df['Ï†ÑÏ≤¥Ïπ®ÏàòÌöüÏàò'] == 0)].copy()

# Îç∞Ïù¥ÌÑ∞ ÌÉÄÏûÖ Î≥ÄÌôò
df_filtered.loc[:, 'Ï£ºÌñâÍ±∞Î¶¨'] = df_filtered['Ï£ºÌñâÍ±∞Î¶¨'].astype(int)
df_filtered.loc[:, 'Í∞ÄÍ≤©'] = df_filtered['Í∞ÄÍ≤©'].astype(float)

# Ï£ºÌñâÍ±∞Î¶¨ Í∑∏Î£πÏù¥ 400,000km Ïù¥ÌïòÏù∏ Îç∞Ïù¥ÌÑ∞ ÌïÑÌÑ∞ÎßÅ
df_filtered = df_filtered[df_filtered['Ï£ºÌñâÍ±∞Î¶¨'] <= 400000]

# Î™®Îç∏ Ï¢ÖÎ•òÎ≥Ñ ÌèâÍ∑† Í∞ÄÍ≤© Í≥ÑÏÇ∞ Î∞è Ï†ïÎ†¨
grouped = df_filtered.groupby(['Ï£ºÌñâÍ±∞Î¶¨'])['Í∞ÄÍ≤©'].mean().reset_index()

# ÏÑ†Ìòï ÌöåÍ∑Ä Î™®Îç∏ ÏÉùÏÑ± Î∞è ÌïôÏäµ
mileage = grouped['Ï£ºÌñâÍ±∞Î¶¨'].values.reshape(-1, 1)
prices = grouped['Í∞ÄÍ≤©'].values

model = LinearRegression()
model.fit(mileage, prices)

# ÏòàÏ∏°Îêú Í∞ÄÍ≤© Í≥ÑÏÇ∞
predicted_prices = model.predict(mileage)

# ÌöåÍ∑Ä Í≥ÑÏàò Î∞è Ï†àÌé∏
slope = model.coef_[0]  # Ï£ºÌñâÍ±∞Î¶¨ 1kmÎãπ Í∞ÄÍ≤© Î≥ÄÌôîÎüâ (ÎßåÏõê)
intercept = model.intercept_

# Ï¥àÍ∏∞ Í∞ÄÍ≤© ÏÑ§Ï†ï (Ï£ºÌñâÍ±∞Î¶¨=0Ïùº ÎïåÏùò Í∞ÄÍ≤©)
initial_price = intercept

# Ï£ºÌñâÍ±∞Î¶¨ 10,000kmÎãπ Í∞êÏÜåÏú® Í≥ÑÏÇ∞
decrease_per_10000km = slope * 10000
percentage_decrease_per_10000km = (decrease_per_10000km / initial_price) * 100

# Í≤∞Í≥º Ï∂úÎ†•
print(f"Ï£ºÌñâÍ±∞Î¶¨ 1kmÎãπ ÌèâÍ∑† Í∞ÄÍ≤© Í∞êÏÜåÎüâ: {slope:.2f} ÎßåÏõê")
print(f"Ï£ºÌñâÍ±∞Î¶¨ 10,000kmÎãπ ÌèâÍ∑† Í∞ÄÍ≤© Í∞êÏÜåÎüâ: {decrease_per_10000km:.2f} ÎßåÏõê")
print(f"Ï£ºÌñâÍ±∞Î¶¨ 10,000kmÎãπ ÌèâÍ∑† Í∞êÏÜåÏú®: {percentage_decrease_per_10000km:.2f}%")

# Í∑∏ÎûòÌîÑ Í∑∏Î¶¨Í∏∞
plt.figure(figsize=(12, 8))

# ÏÇ∞Ï†êÎèÑ Í∑∏Î¶¨Í∏∞ (ÏõêÏùò ÌÅ¨Í∏∞ ÌÇ§Ïö∞Í∏∞)
plt.scatter(grouped['Ï£ºÌñâÍ±∞Î¶¨'], grouped['Í∞ÄÍ≤©'], color='#1f77b4', edgecolor='black', alpha=0.7, s=200)

# ÏÑ†Ìòï ÌöåÍ∑ÄÏÑ† Ï∂îÍ∞Ä (label Ï∂îÍ∞Ä)
plt.plot(grouped['Ï£ºÌñâÍ±∞Î¶¨'], predicted_prices, color='red', linewidth=2.5, label=f'y = {slope:.2f}x + {intercept:.2f}')

# Í∑∏ÎûòÌîÑ ÏÑ§Ï†ï
plt.title("Ï£ºÌñâÍ±∞Î¶¨Ïóê Îî∞Î•∏ Í∞ÄÍ≤© Î∂ÑÌè¨", fontsize=16, pad=20)
plt.xlabel('Ï£ºÌñâÍ±∞Î¶¨ (km)', fontsize=14, labelpad=15)
plt.ylabel('Í∞ÄÍ≤© (ÎßåÏõê)', fontsize=14, labelpad=15)
plt.xlim(0, 400000)
plt.ylim(0, max(prices) * 1.1)

# Î≤îÎ°ÄÎ•º Í∑∏ÎûòÌîÑ Ïïà Ïö∞Ï∏° ÏÉÅÎã®Ïóê Î∞∞Ïπò
plt.legend(loc='upper right', fontsize=12)

plt.tight_layout()

# Í∑∏ÎûòÌîÑÎ•º JPG ÌååÏùºÎ°ú Ï†ÄÏû•
plt.savefig('scatter_price_by_mileage.jpg', format='jpg', dpi=300)

# Í∑∏ÎûòÌîÑ ÌëúÏãú
plt.show()
```

### ¬†Î∞∞ÏßÄ : Ï∞®ÎüâÏùò Îì±Í∏âÏù¥ÎÇò ÏòµÏÖò Ìå®ÌÇ§ÏßÄÎ•º ÏùòÎØ∏ÌïòÎ©∞, Í∏∞Î≥∏Ìòï/ÌîÑÎ¶¨ÎØ∏ÏóÑ/Ïä§Ìè¨Ï∏† Îì±ÏúºÎ°ú ÎÇòÎâ®.

###  ÏÑ∏Î∂Ä Î∞∞ÏßÄ : Î∞∞ÏßÄÎ•º Îçî ÏÑ∏Î∂ÑÌôîÌïú Ï†ïÎ≥¥Î°ú, Î∏îÎûôÏóêÎîîÏÖò, ÏùµÏä§ÌÅ¥Î£®ÏãúÎ∏å Îì± ÌäπÏ†ï Ìä∏Î¶ºÏù¥ÎÇò ÌäπÎ≥Ñ ÏÇ¨ÏñëÏùÑ ÎÇòÌÉÄÎÉÑ. 

# Î∞∞ÏßÄ ÌÅ¥Îü¨Ïä§ÌÑ∞ÎßÅ

```python
#ÏèòÎÇòÌÉÄ
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.cluster import KMeans

# 1) Sonata Ï∂îÏ∂ú
sonata = df[df["Î™®Îç∏Î™Ö"].str.contains("ÏèòÎÇòÌÉÄ")]
Sonata = sonata.groupby('Î∞∞ÏßÄ')['Í∞ÄÍ≤©'].mean().sort_values().reset_index()
Sonata
    
# 3) Í≤∞Ï∏°Ïπò Ï≤òÎ¶¨ (ffill ÏòàÏãú)
Sonata.ffill(inplace=True)

# 4) Ï§ëÎ≥µ Ï†úÍ±∞
Sonata.drop_duplicates(inplace=True)

# 5) ÌäπÏ†ï Ïª¨Îüº Í∏∞Ï§Ä Ï†ïÎ†¨ (Ïòà: ÎßàÏßÄÎßâ ÏàòÏ†ïÏùºÏù¥ Ï°¥Ïû¨ÌïúÎã§Í≥† Í∞ÄÏ†ï)
if "ÎßàÏßÄÎßâ ÏàòÏ†ïÏùº" in Sonata.columns:
    Sonata.sort_values(by="ÎßàÏßÄÎßâ ÏàòÏ†ïÏùº", ascending=True, inplace=True)

# -------------------------------------------------
#   Ïó¨Í∏∞ÏÑúÎ∂ÄÌÑ∞ K-meansÎ•º Ïù¥Ïö©Ìïú Í∞ÄÍ≤©ÎåÄ ÎùºÎ≤®ÎßÅ
# -------------------------------------------------

# A) K-means ÌÅ¥Îü¨Ïä§ÌÑ∞ÎßÅÏóê ÏÇ¨Ïö©Ìï† Îç∞Ïù¥ÌÑ∞(X): Í∞ÄÍ≤©Îßå Îã®Ïùº Ïª¨Îüº ÏÇ¨Ïö©
X = Sonata[['Í∞ÄÍ≤©']]

# B) Íµ∞Ïßë Í∞úÏàò(k) ÏßÄÏ†ï ÌõÑ K-means ÏàòÌñâ
kmeans = KMeans(n_clusters=4, random_state=42)
Sonata['cluster'] = kmeans.fit_predict(X)

# C) Í∞Å Íµ∞ÏßëÏùò Ï§ëÏã¨(ÌèâÍ∑† Í∞ÄÍ≤©) ÌôïÏù∏
centers = kmeans.cluster_centers_
print("ÌÅ¥Îü¨Ïä§ÌÑ∞Î≥Ñ Ï§ëÏã¨(ÌèâÍ∑† Í∞ÄÍ≤©):", centers.ravel())

# D) Íµ∞Ïßë Î≤àÌò∏(cluster)Í∞Ä Í∞ÄÍ≤©ÎåÄ ÏàúÏÑúÎ°ú Ï†ïÎ†¨ÎêòÏßÄ ÏïäÏïòÏùÑ Ïàò ÏûàÏúºÎØÄÎ°ú
#    Ï§ëÏã¨Í∞í(centers)ÏùÑ Í∏∞Ï§ÄÏúºÎ°ú 'Í∞ÄÏû• ÎÇÆÏùÄ Í∞ÄÍ≤©ÎåÄ‚ÜíÍ∞ÄÏû• ÎÜíÏùÄ Í∞ÄÍ≤©ÎåÄ' ÏàúÏÑúÎ°ú Ïû¨Îß§Ìïë
sorted_cluster_idxs = np.argsort(centers.ravel())
cluster_map = {}
for new_label, old_label in enumerate(sorted_cluster_idxs):
    cluster_map[old_label] = new_label

Sonata['cluster_price_order'] = Sonata['cluster'].map(cluster_map)

# E) Í≤∞Í≥º ÌôïÏù∏ÏùÑ ÏúÑÌï¥ Í∞ÄÍ≤© Ïò§Î¶ÑÏ∞®Ïàú Ïû¨Ï†ïÎ†¨
Sonata.sort_values('Í∞ÄÍ≤©', inplace=True)
display(Sonata.head(30))  # Í≤∞Í≥º ÏùºÎ∂Ä ÌôïÏù∏ 
```

## Î∞∞ÏßÄÎ≥Ñ ÎßâÎåÄÍ∑∏ÎûòÌîÑ

#### Î∞∞ÏßÄÍ∞íÏù¥ Ï¶ùÍ∞ÄÌï†ÏàòÎ°ù Í∞ÄÍ≤©ÎèÑ Îã®Í≥ÑÏ†ÅÏúºÎ°ú ÏÉÅÏäπÌïúÎã§

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

plt.figure(figsize=(20, 10))
plt.grid(True, linestyle='--', linewidth=1, alpha=0.5)

# ÎßâÎåÄ Í∑∏ÎûòÌîÑ (x=cluster, y=Í∞ÄÍ≤©, hue=ÏÑ∏Î∂Ä Î∞∞ÏßÄ)
ax = sns.barplot(
    x='cluster_price_order',
    y='Í∞ÄÍ≤©',
    data=Sonata,
    hue='Î∞∞ÏßÄ',
    palette='Blues',
    dodge=False
)

# ------------------------------------------------
# (A) Í∞Å "cluster_price_order"Î≥Ñ ÌèâÍ∑† Í∞ÄÍ≤© Í≥ÑÏÇ∞
# ------------------------------------------------
centers = Sonata.groupby('cluster_price_order')['Í∞ÄÍ≤©'].mean().to_dict()
# Ïòà: {0: 1326.5, 1: 3217.0, 2: 404.8, 3: 2151.8} (Ïã§Ï†ú Í∞íÏùÄ ÏÉÅÌô©Ïóê Îî∞Îùº Îã¨ÎùºÏßà Ïàò ÏûàÏùå)

# xÏ∂ï ÏµúÎåÄÍ∞í (ÌÖçÏä§Ìä∏Î•º Ïò§Î•∏Ï™ΩÏóê ÌëúÏãúÌïòÍ∏∞ ÏúÑÌï¥)
x_min, x_max = ax.get_xlim()

for rank, center_price in centers.items():
    # 1) ÏàòÌèâÏÑ†: Ìï¥Îãπ Îì±Í∏âÏùò ÌèâÍ∑† Í∞ÄÍ≤© ÏúÑÏπòÏóê Ï†êÏÑ†
    plt.axhline(y=center_price, color='blue', linestyle='--', linewidth=2, alpha=0.7)
    
    # 2) ÌÖçÏä§Ìä∏: Í∑∏ÎûòÌîÑ Ïò§Î•∏Ï™Ω ÎÅù Í∑ºÏ≤òÏóê "Îì±Í∏â N = ~~~" ÌëúÏãú
    #    yÍ∞íÏùÄ ÏÑ†Î≥¥Îã§ ÏÇ¥Ïßù ÏúÑÎ°ú (Ïòà: + 10)
    plt.text(
        x=x_max - 3.5, 
        y=center_price + 10,
        s=f"Îì±Í∏â {rank} = {center_price:.1f}",
        color='red',
        fontsize=10
    )

plt.title("<ÏèòÎÇòÌÉÄ : Î∞∞ÏßÄ Í∞ÄÍ≤© Îì±Í∏â>", fontsize=20, weight='bold', pad=20)
plt.xlabel("Î∞∞ÏßÄ", fontsize=18, labelpad=15, weight='bold')
plt.ylabel("Í∞ÄÍ≤©(Îã®ÏúÑ:ÎßåÏõê)", fontsize=18, labelpad=15, weight='bold')
plt.yticks(fontsize=18)
plt.xticks(fontsize=18)

# Î≤îÎ°Ä ÏÑ§Ï†ï
plt.legend(loc='upper left', fontsize=3)

plt.show()
```
